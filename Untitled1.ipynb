```python
!pip install stop-words
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
from wordcloud import WordCloud
from stop_words import get_stop_words

df=pd.read_csv('staff_diplom.csv', delimiter=';')

df3 = df.groupby(['title']).size().reset_index(name='count')
df3=df3.sort_values(by='count')
plt.bar(df3['title'], df3['count'])
plt.show()

df1 = df.groupby(['year']).size().reset_index(name='count')
df1=df1.sort_values(by='count')
plt.bar(df1['year'], df1['count'])
plt.show()

df8=df[['title','year']]
df8=df8.sort_values(by='year')
df8=df8[df8['year']==2003]
plt.show()

df4=df[df['staffid']=='petrosjan']
df4=df[df['department']=='mstnmo']
df4=df['title']
df
df4=pd.DataFrame(df4)
df4=pd.DataFrame.to_string(df4)
STOPWORDS_RU = get_stop_words('russian')
wordcloud = WordCloud(width = 2000, 
                      height = 1500, 
                      random_state=1, 
                      background_color='black', 
                      margin=20, 
                      colormap='Pastel1', 
                      collocations=False, 
                      stopwords = STOPWORDS_RU).generate(df4)
plt.imshow(wordcloud)
plt.show()

df7=df[['department','title']]
df7=df7[df7['department']=='mems']
df7

df6 = df.groupby(['department']).size().reset_index(name='count')
df6=df6.sort_values(by='count')
plt.bar(df6['department'], df6['count'])
plt.show()

df2=df['student']
df
df2=pd.DataFrame(df2)
df2=pd.DataFrame.to_string(df2)
STOPWORDS_RU = get_stop_words('russian')
wordcloud = WordCloud(width = 2000, 
                      height = 1500, 
                      random_state=1, 
                      background_color='black', 
                      margin=20, 
                      colormap='Pastel1', 
                      collocations=False, 
                      stopwords = STOPWORDS_RU).generate(df2)
plt.imshow(wordcloud)
plt.show()

df5=df[['staffid','title']]
df5=df5[df5['staffid']=='petrosjan']
df5
```
